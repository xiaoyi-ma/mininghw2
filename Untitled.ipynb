```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import SpectralClustering
from sklearn import preprocessing

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.cluster.hierarchy as shc
```

# Problem3

# Clustering1

```python
data1 = pd.read_table("C:\\Users\\Dell\\Desktop\\Clustering1.txt")
X1 = np.array(data1)
```

## KMeans

```python
kmeans1 = KMeans(n_clusters=6, random_state=0).fit(X1)
kmeans1.labels_
kmeans1.cluster_centers_
y_kmeans1 = kmeans1.predict(X1)
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(X1[:,0],X1[:,1], c=y_kmeans1, s=25, cmap='viridis')
centers1 = kmeans1.cluster_centers_
plt.scatter(centers1[:,0], centers1[:,1], c='black', s=200, alpha=0.5);
```

## Hierarchical

```python
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend1 = shc.dendrogram(shc.linkage(data1, method='ward'))
```

```python
hier1 = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')  
hier1.fit_predict(data1)
plt.figure(figsize=(10, 7)) 
plt.scatter(X1[:,0],X1[:,1], c=hier1.labels_)
```

## DBSCAN

```python
dbscan1 = DBSCAN(eps=1.7, min_samples = 5)
clusters1 = dbscan1.fit_predict(X1)
plt.figure(figsize=(10, 7)) 
plt.scatter(X1[:, 0], X1[:, 1], c=clusters1, cmap="plasma")
```

# Problem 4

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn import svm
from sklearn.svm import SVC
```

```python
data = pd.read_csv("C:\\Users\\Dell\\Desktop\\SocialNetworkAds.csv")
```

## KNN

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, 2].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)#70% training and 30% test

scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

```python
print("Precision:",metrics.precision_score(y_test, y_pred))
```

```python
print("Recall:",metrics.recall_score(y_test, y_pred))
```

```python
print("f1-score:",metrics.f1_score(y_test, y_pred))
```

## SVM

```python
X1 = data.iloc[:, :-1].values
y1 = data.iloc[:, 2].values
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size = 0.30)
clf = svm.SVC(kernel='rbf')
clf.fit(X_train1, y_train1)
y_pred1 = clf.predict(X_test1)
```

```python
print(confusion_matrix(y_test1,y_pred1))
print(classification_report(y_test1,y_pred1))
```

```python
print("Accuracy:",metrics.accuracy_score(y_test1,y_pred1))
```

```python
print("Precision:",metrics.precision_score(y_test1, y_pred1))
```

```python
print("Recall:",metrics.recall_score(y_test1, y_pred1))
```

```python
print("f1-score:",metrics.f1_score(y_test1, y_pred1))
```

```python
support_vector_indices = clf.support_
print(support_vector_indices)
support_vectors_per_class = clf.n_support_
print(support_vectors_per_class)
```

```python
support_vectors = clf.support_vectors_

plt.scatter(X_train1[:,0], X_train1[:,1])
plt.scatter(support_vectors[:,0], support_vectors[:,1], color='red')
plt.show()
```
